```octave

% Add the readonly directory to path so we can use provided helper functions
% (Like OCVfromSOCtemp.m which exists in your file list)
addpath readonly

% Load the data
load('readonly/pulseData.mat');
load('readonly/pulseModel.mat');

% Initial Setup
T = 25;             % Temperature
deltaT = 1;         % Sample time
Q = getParamESC('QParam',T,model); % Capacity

% Prepare Data Vectors
ik = pulseData.current(:);
vk = pulseData.voltage(:);
tk = pulseData.time(:);

fprintf('Data loaded successfully.\n');
```

```octave
% Cost Function for 3-RC Model
function cost = model_error_3RC(params, ik, v_target, deltaT)
    % params vector: [R0, R1, R2, R3, tau1, tau2, tau3]
    R0 = abs(params(1));
    R = abs(params(2:4));   % [R1, R2, R3]
    tau = abs(params(5:7)); % [tau1, tau2, tau3]

    % Calculate total model voltage drop (Ohmic + 3 Diffusion voltages)
    v_model = R0 * ik; 
    
    % Loop through each of the 3 RC branches
    for j = 1:3
        % Discrete time coefficients for branch j
        % Equation: i_R[k+1] = exp(-dt/tau)*i_R[k] + (1-exp(-dt/tau))*i[k]
        a = exp(-deltaT / tau(j));
        b = 1 - a;
        
        % Filter current to get resistor current for branch j
        i_Rj = filter([0, b], [1, -a], ik);
        
        % Add voltage drop from this branch (v = R * i_R)
        v_model = v_model + R(j) * i_Rj;
    end
    
    % Cost is RMS error between Target (v_measure - OCV) and Model
    % Note: Target is (v_meas - OCV), which equals (v_ohmic + v_diff).
    % Since discharge current is usually negative in these datasets, and voltage drop
    % is negative, we align signs: v_target should match -v_model_drop.
    % To keep it simple: v_cell = OCV - v_drops. 
    % So v_drops_target = OCV - v_meas.
    
    cost = sqrt(mean((v_target - (-v_model)).^2));
end
```

```octave
% 1. Prepare SOC and OCV data
z0 = SOCfromOCVtemp(vk(1), T, model);
zk = z0 - cumsum([0; ik(1:end-1)]) * deltaT / (Q * 3600);
vk_ocv = OCVfromSOCtemp(zk, T, model);

% 2. Isolate the target voltage drop (Polarization)
% v_polarization = v_measured - OCV
v_dyn_target = vk - vk_ocv;

% 3. Set Initial Guesses for 3-RC
% We guess separated time constants to help the optimizer converge
% Params: [R0, R1, R2, R3, tau1, tau2, tau3]
% R0: Ohmic resistance
% Branch 1: Fast (tau ~ 10s)
% Branch 2: Medium (tau ~ 100s)
% Branch 3: Slow (tau ~ 1000s)
x0 = [0.01, ...           % R0
      0.005, 0.005, 0.005, ... % R1, R2, R3 (start small)
      10, 100, 1000];     % tau1, tau2, tau3 (fast, med, slow)

% 4. Run Optimization
cost_func = @(p) model_error_3RC(p, ik, v_dyn_target, deltaT);
options = optimset('Display', 'iter', 'TolFun', 1e-5, 'MaxFunEvals', 5000);

fprintf('Optimizing 3-RC Model (this may take a minute)...\n');
x_opt = fminsearch(cost_func, x0, options);

% 5. Extract Results
R0_opt = abs(x_opt(1));
R_opt = abs(x_opt(2:4));
tau_opt = abs(x_opt(5:7));
C_opt = tau_opt ./ R_opt;

fprintf('\n=== Identified 3-RC Parameters ===\n');
fprintf('R0: %.5f Ohms\n', R0_opt);
for j=1:3
    fprintf('Branch %d: R=%.5f Ohms, C=%.1f F, tau=%.1f s\n', ...
            j, R_opt(j), C_opt(j), tau_opt(j));
end
```

```octave
function [vk_est] = simCell3RC(ik, deltaT, model, z0, R0, R_vec, tau_vec)
    Q = model.QParam;
    n = length(ik);
    
    % Initialize voltage drops
    v_drops = zeros(n, 1);
    
    % 1. Ohmic Drop
    v_drops = v_drops + R0 * ik;
    
    % 2. Loop through 3 branches for Diffusion Drops
    for j = 1:3
        a = exp(-deltaT / tau_vec(j));
        b = 1 - a;
        
        % Filter current to get i_R for this branch
        i_R = filter([0, b], [1, -a], ik);
        
        % Add R*i_R to total drops
        v_drops = v_drops + R_vec(j) * i_R;
    end
    
    % 3. Calculate Final Voltage
    % Simulate SOC
    zk = z0 - cumsum([0; ik(1:end-1)]) * deltaT / (Q * 3600);
    OCV = OCVfromSOCtemp(zk, 25, model);
    
    vk_est = OCV - v_drops;
end
```

```octave
graphics_toolkit("gnuplot");
```

```octave
% Run Validation
vest = simCell3RC(ik, deltaT, model, z0, R0_opt, R_opt, tau_opt);

% Calculate Error
rmse = sqrt(mean((vk - vest).^2)) * 1000; % mV
fprintf('Final 3-RC RMS Error: %.2f mV\n', rmse);

% Plot
figure(2);
subplot(2,1,1);
plot(tk/60, vk, 'b', 'LineWidth', 1.5); hold on;
plot(tk/60, vest, 'r--', 'LineWidth', 1.5);
legend('Measured', '3-RC Model');
xlabel('Time (min)'); ylabel('Voltage (V)');
title(sprintf('3-RC Model Fit (RMSE: %.2f mV)', rmse));
grid on;

subplot(2,1,2);
plot(tk/60, (vk - vest)*1000, 'k');
xlabel('Time (min)'); ylabel('Error (mV)');
title('Prediction Error');
grid on;
```
